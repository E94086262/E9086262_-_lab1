```python
print('welcome to SCISSORS(s),ROCK(r),PAPER(p) game')
import random  #使用亂數
i=1  #i是用來讓後面的while迴圈能夠停止，且要先賦值
counter_tie=0  #用來計算幾次 ties
counter_lose=0  #用來計算幾次 losses
while(i!=0):
    computer = random.choice('rps')  #讓電腦可隨機為 r,p,s
    user1=input('enter your moves:r,p,s')  #user1為玩家本人， 輸入值的部分: r就是石頭，p就是布，s就是剪刀
    if(user1==computer):   #平手的狀況
        if user1=='s':
            print('ROCK VERSUS...')
            print('SCISSORS')
        elif user1=='r':
            print('ROCK VERSUS...')
            print('ROCK')
        else:
            print('ROCK VERSUS...')
            print('PAPER')
        print("It's a tie")
        counter_tie+=1
    elif user1 == 's' and computer == 'r':  #非平手狀況，像這一行的話，玩家出剪刀，電腦出石頭，玩家輸了，後面依此類推
        print('ROCK VERSUS...')
        print('ROCK')
        print('YOU LOSE')
        counter_lose+=1
    elif user1 == 's' and computer == 'p':
        print('ROCK VERSUS...')
        print('PAPER')
        print('YOU WIN')
        i=0
    elif user1 == 'r' and computer == 's':
        print('ROCK VERSUS...')
        print('SCISSORS')
        print('YPU WIN')
        i=0
    elif user1 == 'r' and computer == 'p':
        print('ROCK VERSUS...')
        print('PAPER')
        print('YOU LOSE')
        counter_lose+=1
    elif user1 == 'p' and computer == 's':
        print('ROCK VERSUS...')
        print('SCISSORS')
        print('YOU LOSE')
        counter_lose+=1
    elif user1 == 'p' and computer == 'r':
        print('ROCK VERSUS...')
        print('ROCK')
        print('YOU WIN')
        i=0
print('you have %d ties and %d losses before your first win.' %(counter_tie,counter_lose),end='')
```
